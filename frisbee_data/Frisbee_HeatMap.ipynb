```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde


%matplotlib inline
from numpy import genfromtxt
locations = genfromtxt('activity_1558110975.csv', delimiter=',')
```

```python
x=locations[:,0].T
y=locations[:,1].T

plt.scatter(x, y, s=70, alpha=0.08)
plt.xlim(x.min(),x.max())
plt.ylim(y.min(),y.max())
plt.show()
```

```python
# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=50, edgecolor='',marker = 'h')
plt.xlim(x.min(),x.max())
plt.ylim(y.min(),y.max())
plt.show()
```

```python
from matplotlib.colors import LogNorm
plt.hist2d(x, y, bins=80, norm=LogNorm())
plt.colorbar()
plt.show()
```
